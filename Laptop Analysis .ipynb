#### Processer Generation

```python
import re

def infer_gen_from_ghz(ghz):
    ghz = float(ghz)
    if ghz <= 2.0:
        return "5 Gen"
    elif 2.0 < ghz <= 2.5:
        return "6 Gen"
    elif 2.5 < ghz <= 2.8:
        return "8 Gen"
    elif 2.8 < ghz <= 3.1:
        return "9 Gen"
    elif 3.1 < ghz <= 3.3:
        return "11 Gen"
    elif ghz > 3.3:
        return "13 Gen"
    else:
        return "Unknown"

def extract_gen_or_estimate(cpu):
    if not isinstance(cpu, str):
        return "Unknown"

    # Try extracting gen from model number
    match = re.search(r'[iI][3579]\s?-?\s?(\d{4,5})', cpu)
    if match:
        number = match.group(1)
        if len(number) == 4:
            return f"{int(number[0])} Gen"  # e.g., 7200 -> 7 Gen
        elif len(number) == 5:
            return f"{int(number[:2])} Gen"  # e.g., 1135G7 -> 11 Gen

    # Fallback: Estimate from GHz
    freq_match = re.search(r'(\d\.\d+)\s?GHz', cpu)
    if freq_match:
        return infer_gen_from_ghz(freq_match.group(1))

    return "Unknown"

# Apply to DataFrame
Laptop_df['CPU_Generation'] = Laptop_df['Cpu'].astype(str).apply(extract_gen_or_estimate)

# Sample output
print(Laptop_df[['Cpu', 'CPU_Generation']].head())
```

#### Memory Features

```python
# Flag SSD presence
Laptop_df['Has_SSD'] = Laptop_df['Memory'].astype(str).apply(lambda x: 1 if 'SSD' in x else 0)

# Total memory in GB
def extract_total_memory(memory):
    nums = re.findall(r'(\d+)', str(memory))
    return sum([int(n) for n in nums]) if nums else None

Laptop_df['Total_Memory_GB'] = Laptop_df['Memory'].apply(extract_total_memory)
Laptop_df['Total_Memory_GB']
```

### Z - Test

### Hypothesis Testing and Inference

```python
import pandas as pd
from scipy.stats import ttest_ind, f_oneway
```

```python
# 1. RAM: High (>=8) vs Low (<8)

# Split the data based on RAM size and drop missing prices
low_ram = Laptop_df[Laptop_df['Ram'] < 8]['Price'].dropna()
high_ram = Laptop_df[Laptop_df['Ram'] >= 8]['Price'].dropna()

# Perform independent t-test with unequal variances (Welch's t-test)
tstat_ram, pval_ram = ttest_ind(high_ram, low_ram, equal_var=False)

print("\n")  
print(f"RAM >= 8GB vs < 8GB (Price) - T-statistic: {tstat_ram:.4f}, P-value: {pval_ram:.4f}")
if pval_ram < 0.05:
    print("Reject the null hypothesis: Laptops with higher RAM have significantly different prices.")
else:
    print("Fail to reject the null hypothesis: No significant difference in prices based on RAM size.")
print("\n")  
```

```python
# First, check if the 'Has_SSD' column exists.
if 'Has_SSD' not in Laptop_df.columns:
    print("'Has_SSD' column not found in DataFrame.")
else:
    # Convert 'Has_SSD' to boolean if it's not already.
    Laptop_df['Has_SSD'] = Laptop_df['Has_SSD'].astype(bool)

    # 2. SSD vs HDD - Price comparison
    ssd_prices = Laptop_df[Laptop_df['Has_SSD']]['Price'].dropna()
    hdd_prices = Laptop_df[~Laptop_df['Has_SSD']]['Price'].dropna()
    tstat_ssd, pval_ssd = ttest_ind(ssd_prices, hdd_prices, equal_var=False)

    print("\n")  
    print(f"SSD vs HDD (Price) - T-statistic: {tstat_ssd:.4f}, P-value: {pval_ssd:.4f}")
    if pval_ssd < 0.05:
        print("Reject the null hypothesis: Laptops with SSDs have significantly different prices.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in prices based on SSD vs HDD.")
    print("\n")
```

```python
# 3. Touchscreen vs Non-Touchscreen
Laptop_df['Touchscreen'] = Laptop_df['ScreenResolution'].astype(str).str.contains('Touchscreen', case=False)
touch_prices = Laptop_df[Laptop_df['Touchscreen']]['Price'].dropna()
non_touch_prices = Laptop_df[~Laptop_df['Touchscreen']]['Price'].dropna()
tstat_touch, pval_touch = ttest_ind(touch_prices, non_touch_prices, equal_var=False)

print("\n")  
print(f"Touchscreen vs Non-Touchscreen (Price) - T-statistic: {tstat_touch:.4f}, P-value: {pval_touch:.4f}")
if pval_touch < 0.05: 
    print("Reject the null hypothesis: Touchscreen laptops have significantly different prices.")
else:     
    print("Fail to reject the null hypothesis: No significant difference in prices between touchscreen and non-touchscreen laptops.")
print("\n")    
```

```python
# 4. Windows vs Others
Laptop_df['Windows'] = Laptop_df['OpSys'].astype(str).str.contains('Windows', case=False)
windows_prices = Laptop_df[Laptop_df['Windows']]['Price'].dropna()
non_windows_prices = Laptop_df[~Laptop_df['Windows']]['Price'].dropna()
tstat_windows, pval_windows = ttest_ind(windows_prices, non_windows_prices, equal_var=False)

print("\n")  
print(f"Windows vs Others (Price) - T-statistic: {tstat_windows:.4f}, P-value: {pval_windows:.4f}")
if pval_windows < 0.05:
    print("Reject the null hypothesis: Windows laptops have significantly different prices.")
else:
    print("Fail to reject the null hypothesis: No significant difference in prices between Windows and other OS.")
print("\n")  
```

```python
# 5. Brand vs Price (ANOVA)
brand_groups = [group['Price'].dropna() for _, group in Laptop_df.groupby('Company')]
fstat_brand, pval_brand = f_oneway(*brand_groups)

print("\n")  
print(f"Brand (ANOVA on Price) - F-statistic: {fstat_brand:.4f}, P-value: {pval_brand:.4f}")
if pval_brand < 0.05:
    print("Reject the null hypothesis: There are significant differences in prices between brands.")
else:
    print("Fail to reject the null hypothesis: No significant differences in prices between brands.")
print("\n")  
```

```python
# 6. CPU Brand: Intel vs AMD
Laptop_df['CPU_Brand'] = Laptop_df['Cpu'].astype(str).apply(lambda x: 'Intel' if 'Intel' in x else ('AMD' if 'AMD' in x else 'Other'))
intel_prices = Laptop_df[Laptop_df['CPU_Brand'] == 'Intel']['Price'].dropna()
amd_prices = Laptop_df[Laptop_df['CPU_Brand'] == 'AMD']['Price'].dropna()
tstat_cpu, pval_cpu = ttest_ind(intel_prices, amd_prices, equal_var=False)

print("\n")  
print(f"Intel vs AMD CPU (Price) - T-statistic: {tstat_cpu:.4f}, P-value: {pval_cpu:.4f}")
if pval_cpu < 0.05:
    print("Reject the null hypothesis: Laptops with Intel CPUs have significantly different prices compared to AMD.")
else:
    print("Fail to reject the null hypothesis: No significant difference in prices between Intel and AMD CPUs.")
print("\n")      
```

```python
# 7. GPU: Dedicated vs Integrated
Laptop_df['Dedicated_GPU'] = Laptop_df['Gpu'].astype(str).apply(lambda x: 'Nvidia' in x or 'AMD' in x)
dedicated_prices = Laptop_df[Laptop_df['Dedicated_GPU']]['Price'].dropna()
integrated_prices = Laptop_df[~Laptop_df['Dedicated_GPU']]['Price'].dropna()
tstat_gpu, pval_gpu = ttest_ind(dedicated_prices, integrated_prices, equal_var=False)

print("\n")  
print(f"Dedicated vs Integrated GPU (Price) - T-statistic: {tstat_gpu:.4f}, P-value: {pval_gpu:.4f}")
if pval_gpu < 0.05:
    print("Reject the null hypothesis: Laptops with dedicated GPUs have significantly different prices compared to integrated GPUs.")
else:
    print("Fail to reject the null hypothesis: No significant difference in prices based on GPU type.")
print("\n")      
```

```python
# 8. Screen Size vs Weight
Laptop_df['Screen_Size_Category'] = Laptop_df['Inches'].apply(lambda x: 'Small' if x < 14 else 'Large')
small_weight = Laptop_df[Laptop_df['Screen_Size_Category'] == 'Small']['Weight'].dropna()
large_weight = Laptop_df[Laptop_df['Screen_Size_Category'] == 'Large']['Weight'].dropna()
tstat_weight, pval_weight = ttest_ind(large_weight, small_weight, equal_var=False)

print("\n")  
print(f"Large vs Small Screen (Weight) - T-statistic: {tstat_weight:.4f}, P-value: {pval_weight:.4f}")
if pval_weight < 0.05:
    print("Reject the null hypothesis: Large screen laptops have significantly different weights compared to small screen laptops.")
else:
    print("Fail to reject the null hypothesis: No significant difference in weight based on screen size.")
print("\n")      
```
